```python
import json

#Lectura fichero sentimientos: Sentimientos.txt
feelingsFileName = input("Introduzca el nombre del fichero que contiene los sentimientos de los terminos: ")
sentimientos = open(feelingsFileName)
valores={}
for linea in sentimientos:
    termino, valor = linea.split("\t")
    valores[termino] = int(valor)
print("Fichero de sentimientos leido con exito")
print()

#Lectura fichero tweets: salida_tweets.txt
tweetsFileName = input("Introduzca el nombre del fichero que contiene los tweets: ")
with open(tweetsFileName) as tweetsFile:
    print("Fichero de tweets leido con exito")
    print()
    print("Evaluando palabras de tweets no encontradas en sentimientos")
    for tweetLine in tweetsFile:
        tweetJson = json.loads(tweetLine)
        if "delete" not in tweetJson:
            tweetText = tweetJson["text"]
            print(tweetText)
            
            tweetWords = tweetText.split()            
            count = 0
            notEvaluatedWords = []
            for word in tweetWords:
                lowerWord = word.strip(".,;:!#()?\" ").lower()
                if lowerWord in valores:
                    count = count + valores[lowerWord]
                elif lowerWord.isalpha() and lowerWord != "rt":
                    notEvaluatedWords.append(lowerWord)
            
            for notEvaluatedWord in notEvaluatedWords:
                print(notEvaluatedWord, ":", count)
            print()
```
